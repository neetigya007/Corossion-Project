```python
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from statsmodels.regression.linear_model import OLS

#Load and clean data
df_filtered= df[['Temp', 'Ecorr', 'EcorrNew']].dropna()
df_filtered = df_filtered.sort_values(by='Temp')

# ADF Test Function
def adf_test(series, label):
    result = adfuller(series)
    print(f"\nADF Test for {label}")
    print(f"ADF Statistic: {result[0]}")
    print(f"P-value: {result[1]}")
    if result[1] <= 0.05:
        print(f"{label} is stationary (reject H0)")
        return True
    else:
        print(f"{label} is non-stationary (fail to reject H0)")
        return False
#Manual DF GLS Test Function
def df_gls_manual(series, label, max_lag=1):
    series= series.dropna().astype(float)
    T= len(series)
    trend = np.arrange(1,T+1)
    X= np.column_stack([np.ones(T), trend])
    rho= 1 - (7.0/T)

    y_adj= series.values[1:] - rho*series.values[:-1]
    X_adj = X[1:] - rho * X[:-1]

    model = OLS(y_adj, X_adj).fit()
    residuals = model.resid

    adf_result = adfuller(residuals, maxlag=max_lag, regression='n', autolag=None)
    test_stat, p_value = adf_result[0], adf_result[1]

    print(f"\n[Manual DF-GLS] Test for {label}")
    print(f"Test Statistic: {test_stat}")
    print(f"P-value: {p_value}")
    if p_value <= 0.05:
        print(f"{label} is stationary (reject H0)")
        return True
    else:
        print(f"{label} is non-stationary (fail to reject H0)")
        return False
    # Linear Regression Training & Testing
def train_test_linear_regression(X, y, label):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    X_train = sm.add_constant(X_train)
    X_test = sm.add_constant(X_test)
    
    model = sm.OLS(y_train, X_train).fit()
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    
    r2_train = model.rsquared
    r2_test = r2_score(y_test, y_pred_test)
    
    print(f"\nLinear Regression Results for {label}:")
    print(model.summary())
    print(f"Training R²: {r2_train:.4f}, Testing R²: {r2_test:.4f}")
    
    plt.scatter(X_train['Temp'], y_train, label=f"{label} (Train)", alpha=0.6)
    plt.scatter(X_test['Temp'], y_test, label=f"{label} (Test)", color='red', alpha=0.6)
    plt.plot(X_train['Temp'], y_pred_train, label=f"{label} Regression (R²={r2_train:.4f})", linestyle="--")
    plt.xlabel("Temperature (K)")
    plt.ylabel("Ecorr (mV)")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return r2_train, r2_test

# Run Tests and Regression
r2_values = {}

# Ecorr Blank
if adf_test(df_filtered['Ecorr'], 'Ecorr Blank'):
    r2_values['Blank'] = train_test_linear_regression(df_filtered[['Temp']], df_filtered['Ecorr'], 'Ecorr Blank')

# Ecorr 500 ppm
if adf_test(df_filtered['EcorrNew'], 'Ecorr 500 ppm') or df_gls_manual(df_filtered['EcorrNew'], 'Ecorr 500 ppm'):
    r2_values['500ppm'] = train_test_linear_regression(df_filtered[['Temp']], df_filtered['EcorrNew'], 'Ecorr 500 ppm')

# Final R² Summary
print("\nR² Values Summary:", r2_values)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# -------------------------
# Load and prepare dataset
# -------------------------
df = pd.read_csv("C:/Users/neeti/OneDrive/Desktop/Starch/Expanded_Starch_Dataset.csv")
df = df[['Temp', 'Flowrate', 'Ecorr', 'EcorrNew', 'Icorr', 'IcorrNew']].dropna()

X = df[['Temp', 'Flowrate']]
y_dict = {
    "Ecorr (Blank)": df['Ecorr'],
    "Ecorr (500ppm)": df['EcorrNew'],
    "Icorr": df['Icorr'],
    "IcorrNew": df['IcorrNew']
}

# -------------------------
# Train-test split
# -------------------------
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
train_indices = X_train.index
test_indices = X_test.index

y_train_dict = {k: v.loc[train_indices] for k, v in y_dict.items()}
y_test_dict = {k: v.loc[test_indices] for k, v in y_dict.items()}

# -------------------------
# Hyperparameter grid
# -------------------------
param_dist = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}

# -------------------------
# Optimization function
# -------------------------
def train_and_evaluate(X_train, y_train, X_test, y_test, label):
    rf = RandomForestRegressor(random_state=42)
    random_search = RandomizedSearchCV(
        rf, param_distributions=param_dist,
        n_iter=20, cv=3, n_jobs=-1, random_state=42,
        scoring='neg_mean_squared_error'
    )
    random_search.fit(X_train, y_train)
    best_model = random_search.best_estimator_

    y_pred = best_model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    mad = np.mean(np.abs(y_test - y_pred))
    r2 = r2_score(y_test, y_pred)

    print(f"\n[{label}]")
    print(f"Best Parameters: {random_search.best_params_}")
    print(f"MSE:   {mse:.5f}")
    print(f"RMSE:  {rmse:.5f}")
    print(f"MAE:   {mae:.5f}")
    print(f"MAD:   {mad:.5f}")
    print(f"R²:    {r2:.5f}")

    return y_test.values, y_pred, label

# -------------------------
# Train and evaluate all models
# -------------------------
results = []
for label in y_dict:
    results.append(train_and_evaluate(X_train, y_train_dict[label], X_test, y_test_dict[label], label))


import numpy as np
import matplotlib.pyplot as plt

# Extract results from your trained models
ecorr_blank = [res for res in results if "Ecorr (Blank)" in res[2]][0]
ecorr_500ppm = [res for res in results if "Ecorr (500ppm)" in res[2]][0]

icorr = [res for res in results if res[2] == "Icorr"][0]
icorrnew = [res for res in results if res[2] == "IcorrNew"][0]

# --- Ecorr Plot (Blank and 500ppm) ---

# Sort actual values and predicted for better visual clarity
sorted_idx_blank = np.argsort(ecorr_blank[0])
sorted_idx_500ppm = np.argsort(ecorr_500ppm[0])

actual_blank = np.array(ecorr_blank[0])[sorted_idx_blank]
pred_blank = np.array(ecorr_blank[1])[sorted_idx_blank]

actual_500ppm = np.array(ecorr_500ppm[0])[sorted_idx_500ppm]
pred_500ppm = np.array(ecorr_500ppm[1])[sorted_idx_500ppm]

plt.figure(figsize=(12, 6))
plt.plot(actual_blank, label="Actual Ecorr (Blank)", color='blue', linestyle='-')
plt.plot(pred_blank, label="Predicted Ecorr (Blank)", color='blue', linestyle='--')

plt.plot(actual_500ppm, label="Actual Ecorr (500ppm)", color='green', linestyle='-')
plt.plot(pred_500ppm, label="Predicted Ecorr (500ppm)", color='green', linestyle='--')

plt.title("Line Plot of Actual vs Predicted Ecorr Values (Random Forest)")
plt.xlabel("Test Sample Index (Sorted by Actual Value)")
plt.ylabel("Ecorr (mV)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# --- Icorr and IcorrNew Plot ---

sorted_idx_icorr = np.argsort(icorr[0])
sorted_idx_icorrnew = np.argsort(icorrnew[0])

actual_icorr = np.array(icorr[0])[sorted_idx_icorr]
pred_icorr = np.array(icorr[1])[sorted_idx_icorr]

actual_icorrnew = np.array(icorrnew[0])[sorted_idx_icorrnew]
pred_icorrnew = np.array(icorrnew[1])[sorted_idx_icorrnew]

plt.figure(figsize=(12, 6))
plt.plot(actual_icorr, label="Actual Icorr", color='purple', linestyle='-')
plt.plot(pred_icorr, label="Predicted Icorr", color='purple', linestyle='--')

plt.plot(actual_icorrnew, label="Actual IcorrNew", color='cyan', linestyle='-')
plt.plot(pred_icorrnew, label="Predicted IcorrNew", color='cyan', linestyle='--')

plt.title("Line Plot of Actual vs Predicted Icorr and IcorrNew Values")
plt.xlabel("Test Sample Index (Sorted by Actual Value)")
plt.ylabel("Icorr Value")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# -------------------------
# Scatter Plots: Ecorr and Icorr groups
# -------------------------
ecorr_results = [res for res in results if "Ecorr" in res[2]]
icorr_results = [res for res in results if "Icorr" in res[2]]

# Ecorr scatter plot
plt.figure(figsize=(8, 6))
for actual, predicted, label in ecorr_results:
    plt.scatter(actual, predicted, alpha=0.6, label=label)

min_val = min([min(r[0]) for r in ecorr_results])
max_val = max([max(r[0]) for r in ecorr_results])
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Ideal')

plt.title("Scatter Plot - Ecorr Actual vs Predicted")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Icorr scatter plot
plt.figure(figsize=(8, 6))
for actual, predicted, label in icorr_results:
    plt.scatter(actual, predicted, alpha=0.6, label=label)

min_val = min([min(r[0]) for r in icorr_results])
max_val = max([max(r[0]) for r in icorr_results])
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Ideal')

plt.title("Scatter Plot - Icorr Actual vs Predicted")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
